Observations:
Date Columns: (Date of Purchase, Date of Journey) should be converted to datetime format.

Time Columns: (Time of Purchase, Departure Time, Arrival Time, Actual Arrival Time) should be converted to time format.

Missing Values:
Actual Arrival Time: 1,880 missing values.

Reason for Delay: Only 4,172 non-null values (most are missing).

Potential Analysis Areas:

```python
#Delete last row because it is the same as the column names
rail = rail[rail.index != rail.last_valid_index()]  
```

```python
#  (Date of Purchase, Date of Journey)  converted to datetime format.

rail[['Date of Purchase', 'Date of Journey']] =rail[['Date of Purchase', 'Date of Journey']].apply(pd.to_datetime, errors='coerce')

```

```python
# ( Arrival Time, Actual Arrival Time)  converted to time format.

rail['Actual Arrival Time'] = rail['Actual Arrival Time'].fillna("00:00:00")
rail["Actual Arrival Time"] = pd.to_datetime(rail["Date of Journey"].astype(str) + " " + rail["Actual Arrival Time"].astype(str), errors='coerce')
rail["Arrival Time"] = pd.to_datetime(rail["Date of Journey"].astype(str) + " " + rail["Arrival Time"].astype(str), errors='coerce')
rail['Delay Time'] = (rail['Actual Arrival Time'] - rail['Arrival Time']).dt.total_seconds() / 60
```

```python

rail["Price"] = pd.to_numeric(rail["Price"], errors='coerce')
rail["Refund Request"] = rail["Refund Request"].str.strip().str.capitalize()
rail["Date of Purchase"] = pd.to_datetime(rail["Date of Purchase"], errors='coerce')
rail["Date of Journey"] = pd.to_datetime(rail["Date of Journey"], errors='coerce')
rail["Time of Purchase"] = pd.to_datetime(rail["Time of Purchase"], format='%H:%M:%S', errors='coerce').dt.time
rail["Departure Time"] = pd.to_datetime(rail["Departure Time"], format='%H:%M:%S', errors='coerce').dt.time
```

```python
rail['Delay Time'] = rail.apply(lambda row: 0 if row['Journey Status'] == 'Cancelled' else row['Delay Time'], axis=1)
rail['Delay Time'] = rail['Delay Time'].apply(lambda x: max(x, 0))
```

```python
rail['Reason for Delay'] = rail['Reason for Delay'].replace({
    'Weather Conditions': 'Weather',
    'Signal failure': 'Signal Failure' ,
    'Staffing': 'Staff Shortage'})
```

```python
rail.isnull().sum()
```

## **Missing Values**

```python
rail['Railcard'].unique()
```

```python
rail["Railcard"].fillna('None', inplace=True)
rail["Reason for Delay"].fillna('No Delay', inplace=True)
```

```python
rail["Booking Lead Time"] = (rail["Date of Journey"] - rail["Date of Purchase"]).dt.days
```

# **Model**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.tsa.arima.model import ARIMA
from datetime import datetime, timedelta


daily_rides = rail.groupby('Date of Journey').size().reset_index(name='Ride Count')
daily_rides.set_index('Date of Journey', inplace=True)



train_size = int(len(daily_rides) * 0.8)
train, test = daily_rides.iloc[:train_size], daily_rides.iloc[train_size:]



plt.figure(figsize=(10, 6))
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.title('Historical Daily Ride Counts')
plt.xlabel('Date')
plt.ylabel('Ride Count')
plt.legend()
plt.show()
```

```python
arima_model = ARIMA(train, order=(5, 1, 0))
arima_fit = arima_model.fit()
```

```python
arima_forecast = arima_fit.forecast(steps=len(test))
plt.figure(figsize=(10, 6))
plt.plot(test.index, test, label='Actual')
plt.plot(test.index, arima_forecast, label='ARIMA Forecast', color='red')
plt.title('ARIMA Forecast vs Actual')
plt.xlabel('Date')
plt.ylabel('Ride Count')
plt.legend()
plt.show()
```

```python
arima_rmse = np.sqrt(mean_squared_error(test, arima_forecast))
arima_mae = mean_absolute_error(test, arima_forecast)
print(f'ARIMA RMSE: {arima_rmse}, ARIMA MAE: {arima_mae}')
```

```python
daily_rides['Day'] = daily_rides.index.day
daily_rides['Month'] = daily_rides.index.month
daily_rides['Day of Week'] = daily_rides.index.dayofweek

# Split data again after feature engineering
X = daily_rides[['Day', 'Month', 'Day of Week']]
y = daily_rides['Ride Count']
X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]

```

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(test.index, y_test, label='Actual')
plt.plot(test.index, rf_predictions, label='RF Forecast', color='green')
plt.title('Random Forest Forecast vs Actual')
plt.xlabel('Date')
plt.ylabel('Ride Count')
plt.legend()
plt.show()
```

```python
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_predictions))
rf_mae = mean_absolute_error(y_test, rf_predictions)
print(f'Random Forest RMSE: {rf_rmse}, Random Forest MAE: {rf_mae}')
```

```python
future_dates = [daily_rides.index[-1] + timedelta(days=x) for x in range(1, 31)]
future_df = pd.DataFrame(future_dates, columns=['Date'])
future_df['Day'] = future_df['Date'].dt.day
future_df['Month'] = future_df['Date'].dt.month
future_df['Day of Week'] = future_df['Date'].dt.dayofweek

future_predictions = rf_model.predict(future_df[['Day', 'Month', 'Day of Week']])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(daily_rides.index, daily_rides['Ride Count'], label='Historical')
plt.plot(future_df['Date'], future_predictions, label='Next Month Forecast', color='purple')
plt.title('Next Month Ride Count Forecast')
plt.xlabel('Date')
plt.ylabel('Ride Count')
plt.legend()
plt.show()
```
